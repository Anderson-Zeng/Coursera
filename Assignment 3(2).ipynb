Dataset consists of latitude, longitude, zip codes.

### API Data:

Data about different venues in different neighborhoods of that specific borough is required for the project. In order to gain that data we will use Foursquare locational information, including venue names, locations, menus and even photos. 

First of all, we should make a list of neighborhoods we need. After that we connect to the Foursquare API to gather information about venues inside every neighborhood, setting the radius to be 100 meters.

The detailed information are as follows:
1. Neighborhood names
2. Neighborhood cordinates(latitude, longitude)
3. Venue names
4. Venue cordinates(latitude, longitude)
5. Venue category
